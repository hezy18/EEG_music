```python
# csv文件数据之间是以','分割，unpack是可选参数，默认为False，表示不转置所读取的数据矩阵
    # skiprows=1 表示跳过第一行，,usecols=(1,2,3,4,5)表示读取列数，不读取第0列
import json
# count
with open('EEG_analysis/psd_filt.json') as f:
    psd = json.load(f)
    
import pandas as pd
label_df = pd.read_csv('cut_data/label.csv','\t',usecols=(1,2,3,4,5,6,7))
```

```python
# for preference
import numpy as np

print(label_df['rating'].value_counts())
preference = list(label_df['rating'])
a_mean = np.mean(preference)
a_var = np.var(preference)
a_std = np.std(preference,ddof=1)
print(a_mean,a_var,a_std)

label={}
for i in range(len(label_df)):
    path = label_df['path'][i]
    label[path]=label_df['rating'][i]
```

```python
# for mood
import numpy as np

valence = list(label_df['mood_valence'])
a_mean = np.mean(valence)
a_var = np.var(valence)
a_std = np.std(valence,ddof=1)
print('valence',a_mean,a_var,a_std)

arousal = list(label_df['mood_arousal'])
a_mean = np.mean(arousal)
a_var = np.var(arousal)
a_std = np.std(arousal,ddof=1)
print('arousal',a_mean,a_var,a_std)


mood={'valence':{},'arousal':{}}

for i in range(len(label_df)):
    path = label_df['path'][i]
    mood['valence'][path]=label_df['mood_valence'][i]
    mood['arousal'][path]=label_df['mood_arousal'][i]
```

```python
# label与psd corr: pearson->spearman
from scipy import stats


pref_list = []
psd_arr = np.zeros((319,25))
dic_pref = sorted(label.items(), key=lambda x: x[1])
i=0
for tup in dic_pref:
    psd_arr[i,:] = np.array(psd[tup[0]])
    pref_list.append(tup[1])
    i+=1
    
print('for preference',len(pref_list),len(list(psd_arr[:,0])))

preference_corr=[]
preference_p=[]
for i in range(25):
    #ttest = stats.ttest_ind(pref_list,list(psd_arr[:,i]))
    #r, p = stats.pearsonr(pref_list,list(psd_arr[:,i]))
    r, p = stats.pearsonr(pref_list,list(psd_arr[:,i]))
    preference_corr.append(r)
    if p <0.05:
        print(i)
        preference_p.append(1)
    else:
        preference_p.append(0)
    print(r,p)

    
val_list = []
psd_arr = np.zeros((319,25))
dic_val = sorted(mood['valence'].items(), key=lambda x: x[1])
#print(dic_val)
i=0
for tup in dic_val:
    #print(psd[tup[0]])
    psd_arr[i,:] = np.array(psd[tup[0]])
    val_list.append(tup[1])
    i+=1
    
print('\nfor valence',len(val_list),len(list(psd_arr[:,0])))

valence_corr=[]
valence_p=[]
for i in range(25):
    #ttest = stats.ttest_ind(val_list,list(psd_arr[:,i]))
    #r, p = stats.pearsonr(val_list,list(psd_arr[:,i]))
    r, p = stats.pearsonr(val_list,list(psd_arr[:,i]))
    valence_corr.append(r)
    if p <0.05:
        print(i)
        valence_p.append(1)
    else:
        valence_p.append(0)
    print(r,p)


aro_list = []
psd_arr = np.zeros((319,25))
dic_aro = sorted(mood['arousal'].items(), key=lambda x: x[1])
i=0
for tup in dic_aro:
    psd_arr[i,:] = np.array(psd[tup[0]])
    aro_list.append(tup[1])
    i+=1
    
print('\nfor arousal',len(aro_list),len(list(psd_arr[:,0])))

arousal_corr=[]
arousal_p=[]
for i in range(25):
    #ttest = stats.ttest_ind(aro_list,list(psd_arr[:,i]))
    #r, p = stats.pearsonr(aro_list,list(psd_arr[:,i]))
    r, p = stats.pearsonr(aro_list,list(psd_arr[:,i]))
    arousal_corr.append(r)
    if p <0.05:
        print(i)
        arousal_p.append(1)
    else:
        arousal_p.append(0)
    print(r,p)
    
```

```python
# 均值检验/方差分析：各类别
from scipy import stats
#ttest = stats.ttest_ind(aro_list,list(psd_arr[:,i]))
#stats.f_oneway
LV_psd=[]
HV_psd=[]
for i in range(len(val_list)):
    val = val_list[i]
    if val<0.5:
        LV_psd.append(list(psd_arr[i,:]))
    
    elif val==0.5:
        print('0.5')
    else:
        HV_psd.append(list(psd_arr[i,:]))
LV = np.array(LV_psd)
HV = np.array(HV_psd)
print(LV.shape[0],HV.shape[0])
for i in range(25):
    print(stats.f_oneway(LV[:,i],HV[:,i]))
    
LA_psd=[]
HA_psd=[]
for i in range(len(aro_list)):
    aro = aro_list[i]
    if aro<0.5:
        LA_psd.append(list(psd_arr[i,:]))
    
    elif aro==0.5:
        print('0.5')
    else:
        HA_psd.append(list(psd_arr[i,:]))
LA = np.array(LA_psd)
HA = np.array(HA_psd)
print(LA.shape[0],HA.shape[0])
for i in range(25):
    print(stats.f_oneway(LA[:,i],HA[:,i]))
```
